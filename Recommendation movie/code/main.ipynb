```python
# So we are going to keep the columns which we required.
# we will remove budget as we dont reccomand movie depending on budget.
# Also we will keep generes as it gives the information about the movie which are action, drama, crime, Animi kind of info.
#keywords does matter
#id is used to fetch the images of the movie from the tmdb website wwhen we will doing web.
# original language is not that important as it just sugget the language of the movie.
#we will keep original titile.
#we dont recommand production companies to friends
#we also do not recommand production countries
#release date can be a factor as people may feel to watch movie of their time or new generation want to watch latest movies depends but as of now are not considering.
#revenue can be a factor,
#we are not considering runtime, tagline, vote avg, movie id, vote average
```

```python
movie=movies[["movie_id","genres","title","overview","keywords","cast","crew"]]
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
        
#so if you see the list they have multiple dictionary in the list and they are acting as string as '' symbol is used so we need to first convert them into list and then we can iterate them.
#So to make them to list we have a function in python called ast.literal_eval()
```

```python
movie["genres"]=movie["genres"].apply(convert)
```

```python
movie
```

```python
movie["keywords"]=movie["keywords"].apply(convert)
```

```python
movie["cast"].iloc[0]
```

```python
movie
```

```python
def covert3(obj):
    a=[i["name"] for i in ast.literal_eval(obj)[:3]]
    return a
        
```

```python
movie["cast"]=movie["cast"].apply(covert3)
```

```python
movie
```

```python
def convert4(obj):
    a=[i["name"] for i in ast.literal_eval(obj) if i["job"]=="Director"]
    return a
```

```python
# # 
movie["crew"]=movie["crew"].apply(convert4)

# # Pretty-print the first row's genres
# print(json.dumps(movies["crew"].iloc[0], indent=4))
```

```python
movie
```

```python
movie["overview"]= movie['overview'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie["cast"]=movie["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movie["keywords"]=movie["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movie["genres"]=movie["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movie["crew"]=movie["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie
```

```python
movie["tag"]=movie["overview"]+ movie["genres"]+ movie["keywords"]+ movie["cast"]+ movie["crew"]
```

```python
movie
```

```python
new_df= movie[["movie_id","title","tag"]]
```

```python
new_df["tag"]=new_df["tag"].apply(lambda x: " ".join(x))
```

```python
new_df["tag"][0]
```

```python
new_df["tag"]= new_df["tag"].apply(lambda x: x.lower())
```

```python
new_df["tag"][0]
```

```python
new_df.head()
```

```python
# Now we have to do text vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000, stop_words="english")
```

```python
cv.fit_transform(new_df["tag"]).toarray().shape
```

```python
vectors=cv.fit_transform(new_df["tag"]).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tag"]= new_df["tag"].apply(stem)
```

```python
new_df["tag"]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index= new_df[new_df["title"]==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```
